## Wavelet normalisation test

Write Morlet wavelet in the real space

```python
ko = 5.336
scl = 10.

x = np.arange(-64,64)

wav = (1./scl) * np.cos(ko*x/scl) * np.exp(-0.5*np.abs(x/scl)**2.)

plot(x,wav)
print x.size
```

Plot the Fourier transform

```python
#k = (x) / 128.
#shiftx = 128/2
#k = np.roll(k, shiftx)
k = np.fft.fftfreq(128)
k = np.fft.fftshift(k)

wavFFT = np.fft.fft(wav)
wavFFT = np.fft.fftshift(wavFFT)

plot(k, np.abs(wavFFT))
plot([0.,0.],[0,1.2],'--')

print "Peak =", 1. / scl

print "Peak = ", np.max(np.abs(wavFFT))
```

Write Morlet wavelet in the Fourier space

```python
sclk = 2*np.pi*scl

wavFT = 2*sclk * np.exp(-0.5*np.abs(sclk*k-ko)**2.)

plot(k,wavFT)
plot([1./scl,1./scl],[0,128],'--')

print "max = ", np.max(wavFT)
```
