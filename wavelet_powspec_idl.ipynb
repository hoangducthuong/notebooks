```idl
fbm = fbm2d(-3.1,256,256,seed)
mamdct,75
window,xsize=800,ysize=800
imaffi,fbm,/bar
```

```idl
powspec_k_nosquare,fbm,1,tab_k,spec_k

fan_trans, fbm, [256,256], 1, wt, wav_k, Sk, S11, apodize=0.98

;mamdct,75
window,xsize=800,ysize=500
plot,tab_k,spec_k,/xlog,/ylog,xtitle='Wavenumber k',ytitle='P[k]'
oplot,wav_k, Sk,psym=4
```
